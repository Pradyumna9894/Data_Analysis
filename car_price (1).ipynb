```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("car_price_dataset.csv")
```

Dataset Description: 

The Car Price Dataset contains 10,000 records with 10 attributes detailing used cars and their resale prices.  The dataset is ideal for price prediction models and market analysis, revealing trends such as depreciation patterns and the rising popularity of hybrid and electric vehicles due to environmental concerns.

Columns in dataset: 

* Brand : The car manufacturer
* Model : The specific model of the car.
* Year: The manufacturing year of the car (range: 2000 to 2023).
* Engine_Size: The size of the engine in liters (range: 1.0L to 5.0L).
* Fuel_Type : The type of fuel used (e.g., Diesel, Hybrid, Electric).
* Transmission: The type of transmission (e.g., Manual, Automatic, Semi-Automatic).
* Mileage: The total distance the car has traveled, in kilometers (range: 25 to 299,947).
* Doors : The number of doors (range: 2 to 5).
* Owner_Count : The number of previous owners (range: 1 to 5).
* Price: The selling price of the car in USD (range: $2,000 to $18,301)

```python
sns.histplot(df['Price'], kde=True, color='skyblue')
plt.title('Distribution of Car Prices')
```

DataSet Observation:
Insights from the Car Price Dataset Visualizations
1. Price Trends by Year
    * The Price vs. Year scatter plot and violin plot reveal that newer cars (post-2015) have significantly higher prices, while       older models tend to be cheaper.
    * The median price steadily increases for recent cars, reflecting their higher market value.
2. Mileage and Price Relationship
    * The Mileage vs. Price scatter plot highlights a clear trend:
    * Cars with higher mileage tend to have lower prices, reflecting wear and depreciation.
    * Lower-mileage cars retain their value better, especially newer models.
3. Engine Size Impact
    * The Engine Size vs. Price and Engine Size vs. Mileage scatter plots show that:
    * Cars with larger engines generally have higher prices, indicating more powerful and premium vehicles.
    * Larger engines tend to have lower mileage, suggesting they may be used less frequently or preserved for special purposes.
